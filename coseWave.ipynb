```python
import numpy as np
import matplotlib.pyplot as plt

f = 2  # frequency in Hz
fs = 100  # sampling frequency in Hz
a = 10  # amplitude
theta = np.deg2rad(30)  # phase in radians

x = np.arange(0, 1, 1/fs)  # time vector from 0 to 1 second
y = a * np.cos(2 * np.pi * f * x + theta)  # cosine wave signal   

plt.plot(x, y)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

f = 2  # frequency in Hz
fs = 100  # sampling frequency in Hz
a = 10  # amplitude

# Generate x values from 0 to 2
x = np.linspace(0, 2 * np.pi, fs)

# Compute sine values
y = a * np.cos(x)

# Plot the sine wave
plt.plot(x, y)
plt.title("Cosine Wave")
plt.xlabel("X")
plt.ylabel("cos(X)")
plt.grid(True)

plt.show()
```
