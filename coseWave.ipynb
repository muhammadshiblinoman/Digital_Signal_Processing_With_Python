```python
import numpy as np
import matplotlib.pyplot as plt

f = 5  # frequency in Hz
fs = 100  # sampling frequency in Hz
x = np.arange(0, 1, 1/fs)  # time vector from 0 to 1 second
y = np.cos(2 * np.pi * f * x)  # cosine wave signal   

plt.plot(x, y)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2
x = np.linspace(0, 2 * np.pi, 500)

# Compute sine values
y = np.cos(x)

# Plot the sine wave
plt.plot(x, y)
plt.title("Cosine Wave")
plt.xlabel("X")
plt.ylabel("cos(X)")
plt.grid(True)

plt.show()
```
