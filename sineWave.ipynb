```python
import numpy as np
import matplotlib.pyplot as plt

f = 5  # frequency in Hz
fs = 100  # sampling frequency in Hz
x = np.arange(0, 1, 1/fs)  # time vector from 0 to 1 second
y = np.sin(2 * np.pi * f * x)  # sine wave signal   

plt.plot(x, y)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2
x = np.linspace(0, 2 * np.pi, 500)

# Compute sine values
y = np.sin(x)

# Plot the sine wave
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.grid(True)

plt.show()
```

#s2(t)=Ae^(j3t)=Acos(3t)+jAsin(3t)

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1                 # amplitude
fs = 200              # sampling frequency
t = np.arange(0, 1, 1/fs)

omega = 3 * np.pi     # angular frequency

# Real parts
real_part = A * np.cos(omega * t)

# Imaginary parts
imag_part = A * np.sin(omega * t)

# Plot real part
plt.figure()
plt.plot(t, real_part)
plt.title("Real Part: A cos(3t)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# Plot imaginary part
plt.figure()
plt.plot(t, imag_part)
plt.title("Imaginary Part: A sin(3t)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
