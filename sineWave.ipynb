```python
import numpy as np
import matplotlib.pyplot as plt

f = 2  # frequency in Hz
fs = 100  # sampling frequency in Hz
a = 10  # amplitude
theta = np.deg2rad(45)  # phase in radians

x = np.arange(0, 1, 1/fs)  # time vector from 0 to 1 second
y = a * np.sin(2 * np.pi * f * x + theta)  # sine wave signal   

plt.plot(x, y)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```
