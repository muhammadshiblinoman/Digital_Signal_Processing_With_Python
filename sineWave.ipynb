```python
# Sine Wave Generator
This notebook generates and plots a sine wave using NumPy and Matplotlib.

Adjust the parameters in the next cell: `fs` (sampling rate), `f` (frequency), and `duration`.
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio

# Parameters (edit these as needed)
fs = 8000       # sampling frequency (Hz)
f = 440         # sine frequency (Hz)
duration = 1.0  # seconds

# Time vector and signal
t = np.arange(0, duration, 1.0/fs)
y = np.sin(2 * np.pi * f * t)

# Plot (show a short time window for clarity)
plt.figure(figsize=(10,4))
plt.plot(t, y)
plt.title(f'Sine wave: {f} Hz')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.xlim(0, min(0.01, duration))
plt.show()

# Play audio (inline)
Audio(y, rate=fs)
```
