```python
import numpy as np
import matplotlib.pyplot as plt

A = 1                 # amplitude
fs = 200              # sampling frequency
t = np.arange(0, 1, 1/fs)

omega = 3 * np.pi     # angular frequency

# Real parts
real_part = A * np.cos(omega * t)

# Imaginary parts
imag_part = A * np.sin(omega * t)

# Plot real part
plt.figure()
plt.plot(t, real_part)
plt.title("Real Part: A cos(3t)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# Plot imaginary part
plt.figure()
plt.plot(t, imag_part)
plt.title("Imaginary Part: A sin(3t)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
